```python
import tkinter as tk
from tkinter import filedialog, messagebox
from tkcalendar import DateEntry
import pandas as pd
import arrow
import calendar


root = tk.Tk()

root.geometry('1920x1080')


# ####################################### Import_Csv ################################################################
main_csv = 0
machine_efficiency = 0


# uploading MC efficiency
def machine():
    global machine_efficiency
    fileopen = filedialog.askopenfile(initialdir='C:/Users/ANIMESH KUMAR/Downloads', title='Select a file',
                                      filetypes=[('csv file', "*.csv")])
    if fileopen is not None:
        machine_efficiency = pd.read_csv(fileopen, dtype='unicode')

# uploading Insta_DBR file


def op():
    global main_csv, stock
    fileopen = filedialog.askopenfile(initialdir='C:/Users/ANIMESH KUMAR/Downloads', title='Select a file',
                                      filetypes=[('csv file', "*.csv")])
    if fileopen is not None:

        main_csv = pd.read_csv(fileopen, dtype='unicode', header=2, parse_dates=['Bill Open Date'], dayfirst=True)

        for key in main_csv['Center Name'].unique():
            list_0.insert('end', key)

        main_csv['month_year'] = pd.to_datetime(main_csv['Bill Open Date']).dt.to_period('M')
        stock = pd.DataFrame(data=main_csv,
                             columns=['Bill Open Date', 'Service Sub Group', 'Center Name', 'Net Amt w/o Taxes'])

    else:
        messagebox.showerror('ERROR', 'Please Upload the file')

#  #######################################End_Import_Csv#################################################################


# ####################################### Frontend_design #############################################################
# flat, groove, raised, ridge, solid, or sunken

top_logo = tk.Label(text='MIS FORCE', fg='dark blue', font=("bold", 70))
top_logo.place(x=450, y=45)
root.title("7Med India")

daily_business_report_button = tk.Button(root, text='Click to Import', width=30, font=14, height=2, fg='white',
                                         bg='#AE0000', command=op)
daily_business_report_button.place(x=350, y=256)

daily_business_report = tk.Label(root, text="Import Daily Business Report:", font=("bold", 14), fg='black')
daily_business_report.place(x=90, y=260)

machine_no_report_button = tk.Button(root, text='Click to Import', width=30, font=14, height=2, fg='white',
                                     bg='#AE0000', command=machine)
machine_no_report_button.place(x=350, y=350)

machine_no_report = tk.Label(root, text="Import Machine Number Report:", font=("bold", 14), fg='black')
machine_no_report.place(x=74, y=350)

select_current_month = tk.Label(root, text="Select Current Month :", font=("bold", 14), fg='black')
select_current_month.place(x=150, y=450)

utc = arrow.utcnow()
date_to = DateEntry(root, width=35, bg="dark blue", fg="white", year=int(utc.format("YYYY")),
                    month=int(utc.format("MM")))
date_to.place(x=350, y=450)

centres = tk.Label(root, text="Select Centres :", width=20, font=("bold", 15))
centres.place(x=700, y=250)

list_0 = tk.Listbox(root, selectmode="multiple", height=26, width=50)
list_0.place(x=900, y=250)

listbox = tk.Listbox(root, height=100,
                     width=100,
                     bg="white",
                     activestyle='dotbox',
                     font="Helvetica",
                     fg="yellow", selectmode="multiple")
# ################################################ End of Frontend code #####################################


def show_output():

    root_1 = tk.Tk()

    root_1.grid_rowconfigure(0, weight=1)
    root_1.columnconfigure(0, weight=1)

    frame_main = tk.Frame(root_1, bg="#26b768")
    frame_main.grid(sticky='news')
    # ################################################## Output GUI #############################################
    # Create a frame for the canvas with non-zero row&column weights
    frame_canvas = tk.Frame(frame_main)
    frame_canvas.grid(row=0, column=0, pady=(10, 100), sticky='nw')
    frame_canvas.grid_rowconfigure(0, weight=1)
    frame_canvas.grid_columnconfigure(0, weight=1)

    # Add a canvas in that frame
    canvas = tk.Canvas(frame_canvas, width=1350, height=200, bg="#6D7B8D")
    canvas.grid(row=1, column=0, sticky="news")
    tk.Label(frame_canvas, text='Treatment', font=('Arial', 10, "bold")).grid(row=0, column=0)

    # Add a canvas in that frame
    canvas_1 = tk.Canvas(frame_canvas, width=1350, height=200, bg="#6D7B8D")
    canvas_1.grid(row=5, column=0, sticky="news")
    tk.Label(frame_canvas, text='Revenue', font=('Arial', 10, "bold")).grid(row=4, column=0)

    # Add a canvas in that frame
    canvas_2 = tk.Canvas(frame_canvas, width=1300, height=150, bg="#6D7B8D")
    canvas_2.grid(row=9, column=0, sticky="news")
    tk.Label(frame_canvas, text='MC Efficiency', font=('Arial', 10, "bold")).grid(row=8, column=0)

    # Link a scrollbar to the canvas
    vsb = tk.Scrollbar(frame_canvas, orient="vertical", command=canvas.yview)
    vsb.grid(row=1, column=1, sticky='ns')
    canvas.configure(yscrollcommand=vsb.set)

    # Link a scrollbar to the canvas
    vsb_1 = tk.Scrollbar(frame_canvas, orient="vertical", command=canvas_1.yview)
    vsb_1.grid(row=5, column=1, sticky='ns')
    canvas_1.configure(yscrollcommand=vsb_1.set)

    # Link a scrollbar to the canvas
    vsb_2 = tk.Scrollbar(frame_canvas, orient="vertical", command=canvas_2.yview)
    vsb_2.grid(row=9, column=1, sticky='ns')
    canvas_2.configure(yscrollcommand=vsb_2.set)

    # Create a horizontal scrollbar linked to the canvas.
    hs_bar = tk.Scrollbar(frame_canvas, orient=tk.HORIZONTAL, command=canvas.xview)
    hs_bar.grid(row=2, column=0, sticky=tk.EW)
    canvas.configure(xscrollcommand=hs_bar.set)

    # Create a horizontal scrollbar linked to the canvas.
    hs_bar_1 = tk.Scrollbar(frame_canvas, orient=tk.HORIZONTAL, command=canvas_1.xview)
    hs_bar_1.grid(row=6, column=0, sticky=tk.EW)
    canvas_1.configure(xscrollcommand=hs_bar_1.set)

    # Create a horizontal scrollbar linked to the canvas.
    hs_bar_2 = tk.Scrollbar(frame_canvas, orient=tk.HORIZONTAL, command=canvas_2.xview)
    hs_bar_2.grid(row=10, column=0, sticky=tk.EW)
    canvas_2.configure(xscrollcommand=hs_bar_2.set)

    # ################################################# End Of GUI #############################################
    # ############################################### Treatment GUI #############################################
    # Create a frame to contain the buttons
    frame_buttons = tk.Frame(canvas, bg="#446ab1")
    canvas.create_window((0, 0), window=frame_buttons, anchor='nw')

    to_var = date_to.get_date()
    date_var = to_var.replace(day=1)
    selected_value = []

    for i in list_0.curselection():
        selected_value.append(list_0.get(i))
        from_date = arrow.get(date_var).shift(days=-2)
        to_date = arrow.get(to_var)
        month_end_value = to_date.ceil('month')

        total_date = (to_date - from_date)

        # ############threemonths_back####################
        three_months = arrow.get(date_var).shift(months=-3)

        month = []
        for n in range(3):
            month.append(three_months.shift(months=n).format("YYYY-MM"))

        # ############################################### Treatment loop ################################
        centres = selected_value
        month = month[::-1]

        for i in range(len(centres)):

            grand_total = 0
            from_date = arrow.get(date_var).shift(days=-2)
            for j in range(total_date.days):

                from_date = from_date.shift(days=1)
                date = from_date.date()
                date_month = from_date.format('DD MMMM')
                weekend = from_date.weekday()
                weekend_1 = calendar.day_name[weekend]

                if j == 0:
                    tk.Label(frame_buttons, text="Centre", font=("Arial", 8), width=10,
                             bg="sky blue").grid(row=0, column=0)
                    date = 0
                else:

                    tk.Label(frame_buttons, text=date_month, font=("Arial", 8), width=9,
                             bg="sky blue").grid(row=0, column=j)
                    tk.Label(frame_buttons, text=weekend_1, font=("Arial", 8), width=9,
                             bg="sky blue").grid(row=1, column=j)
                med_filter = stock[(main_csv['Center Name'] == centres[i]) &
                                   (main_csv['Service Sub Group'] == 'Dialysis') &
                                   (main_csv['Bill Open Date'] == str(date))]

                if j == 0:
                    label_centre = tk.Label(frame_buttons, text=centres[i], width=35, font=("Arial", 8))
                    label_centre.grid(row=i + 2, column=j)

                elif j == j:
                    label_value_treat = tk.Label(frame_buttons, text=str(len(med_filter[['Bill Open Date']])), width=9)
                    label_value_treat.grid(row=i + 2, column=j)

                grand_total = grand_total + len(med_filter[['Bill Open Date']])

                if j == total_date.days - 1:
                    pro_data = grand_total / (total_date.days - 1) * int(month_end_value.format("DD"))
                    pro_data_1 = "{:.2f}".format(pro_data)
                    tk.Label(frame_buttons, text="Grand Total", font=("Arial", 8), width=10,
                             bg="sky blue").grid(row=0, column=j + 1)
                    tk.Label(frame_buttons, text="Pro date", font=("Arial", 8), width=8,
                             bg="sky blue").grid(row=0, column=j + 2)
                    label_value_id = tk.Label(frame_buttons, text=grand_total, width=8)
                    pro_data_id = tk.Label(frame_buttons,
                                           text=pro_data_1, width=8)
                    pro_data_id.grid(row=i + 2, column=j + 2)
                    label_value_id.grid(row=i + 2, column=j + 1)

        # #################################################### Treatment_3months_back ##################################

        if i == len(centres) - 1:
            for z in range(len(month)):

                grand_total = 0

                for a in range(len(centres)):
                    med_filter = stock[
                        (main_csv['Center Name'] == centres[a]) & (main_csv['Service Sub Group'] == 'Dialysis') & (
                                main_csv['month_year'] == str(month[z]))]
                    tk.Label(frame_buttons, text=month[z].format("MMMM"), font=("Arial", 8), width=10,
                             bg="sky blue").grid(row=0,
                                                 column=z + total_date.days + 2)

                    grand_total = grand_total + len(med_filter[['Bill Open Date']])

                    label_three_month = tk.Label(frame_buttons, text=str(len(med_filter[['Bill Open Date']])), width=10)
                    label_three_month.grid(row=a + 2, column=z + total_date.days + 2)

    # Update buttons frames idle tasks to let tkinter calculate buttons sizes
    frame_buttons.update_idletasks()
    canvas.config(scrollregion=canvas.bbox("all"))

    # ############################################ Revenue Tab ##############################################
    # 2 Create a frame to contain the buttons
    frame_buttons_1 = tk.Frame(canvas_1, bg="#1d7e49")
    canvas_1.create_window((0, 0), window=frame_buttons_1, anchor='nw')

    # ############################################## Revenue loop ###########################################

    to_var = date_to.get_date()
    date_var = to_var.replace(day=1)
    selected_value = []

    for i in list_0.curselection():
        selected_value.append(list_0.get(i))
    from_date = arrow.get(date_var).shift(days=-2)
    to_date = arrow.get(to_var)
    month_end_value = to_date.ceil('month')

    total_date = (to_date - from_date)

    # Centres input from the user
    centres = selected_value

    for i in range(len(centres)):

        grand_total = 0
        from_date = arrow.get(date_var).shift(days=-2)
        for j in range(total_date.days):

            from_date = from_date.shift(days=1)
            date = from_date.date()
            date_month = from_date.format('DD MMMM')
            weekend = from_date.weekday()
            weekend_2 = calendar.day_name[weekend]
            if j == 0:
                date = 0
            else:
                tk.Label(frame_buttons_1, text=date_month, font=("Arial", 8), width=9, bg="sky blue").grid(row=0,
                                                                                                           column=j)
                tk.Label(frame_buttons_1, text=weekend_2, font=("Arial", 8), width=9, bg="sky blue").grid(row=1,
                                                                                                          column=j)

            med_filter = stock[
                (main_csv['Center Name'] == centres[i]) & (main_csv['Bill Open Date'] == str(date))]

            if j == 0:
                label_centre_rev = tk.Label(frame_buttons_1, text=centres[i], width=35, font=("Arial", 8))
                label_centre_rev.grid(row=i + 2, column=j)
                tk.Label(frame_buttons_1, text="Centre", font=("Arial", 8), width=10, bg="sky blue").grid(row=0,
                                                                                                          column=0)

            elif j == j:
                label_value_rev = tk.Label(
                    frame_buttons_1, text=pd.to_numeric(med_filter['Net Amt w/o Taxes'],
                                                        errors='coerce').sum(), width=10)
                label_value_rev.grid(row=i + 2, column=j)

                grand_total = grand_total + pd.to_numeric(med_filter['Net Amt w/o Taxes'], errors='coerce').sum()

            if j == total_date.days - 1:
                tk.Label(frame_buttons_1, text="Grand Total", font=("Arial", 8), width=10, bg="sky blue").grid(
                    row=0, column=j + 1)
                tk.Label(frame_buttons_1, text="Pro Data", font=("Arial", 8), width=8, bg="sky blue").grid(
                    row=0, column=j + 2)
                pro_data = grand_total / (total_date.days - 1) * int(month_end_value.format("DD"))
                pro_data_1 = "{:.2f}".format(pro_data)

                revenue_pro_data = tk.Label(frame_buttons_1, text=pro_data_1, width=15)
                revenue_pro_data.grid(row=i + 2, column=j + 2)
                label_grand_total = tk.Label(frame_buttons_1, text=grand_total, width=8)
                label_grand_total.grid(row=i + 2, column=j + 1)

    # #################################################### Revenue_3months_back ##################################
    try:
        if i == len(centres) - 1:
            for z in range(len(month)):
                grand_total = 0

                for a in range(len(centres)):
                    med_filter = stock[
                        (main_csv['Center Name'] == centres[a]) & (main_csv['month_year'] == str(month[z]))]
                    tk.Label(frame_buttons_1, text=month[z].format("MMMM"), font=("Arial", 8), width=10,
                             bg="sky blue").grid(row=0, column=z + total_date.days + 2)

                    grand_total = grand_total + len(med_filter[['Bill Open Date']])

                    label_three_month = tk.Label(frame_buttons_1,
                                          text=str(
                                              pd.to_numeric(med_filter['Net Amt w/o Taxes'], errors='coerce').sum()),
                                          width=10)
                    label_three_month.grid(row=a + 2, column=z + total_date.days + 2)

    except:
        pass

    # 2 Update buttons frames idle tasks to let tkinter calculate buttons sizes
    frame_buttons_1.update_idletasks()
    canvas_1.config(scrollregion=canvas_1.bbox("all"))

    # ################################## Machine Efficiency tab ##################################

    frame_buttons_2 = tk.Frame(canvas_2, bg="#C35817")
    canvas_2.create_window((0, 0), window=frame_buttons_2, anchor='nw')

    # ################################# Machine Efficiency #######################################
    to_date = arrow.get(to_var)
    from_date = to_date.replace(day=1).shift(days=-2)
    total_date = (to_date - from_date)

    centres = selected_value
    sum_data = 0
    for i in range(len(centres)):

        from_date = to_date.replace(day=1).shift(days=-2)
        sum_data = sum_data * 0
        grand_total = 0
        for j in range(total_date.days):

            from_date = from_date.shift(days=1)
            date = from_date.date()
            date_month = from_date.format('DD MMMM')

            try:

                if centres[i] in machine_efficiency['Centres'].values:
                    machine_efficiency_value = machine_efficiency.loc[machine_efficiency['Centres'] == centres[i],
                                                                      'Machine No'].iloc[0]
                    tk.Label(frame_buttons_2, text=date_month, font=("Arial", 8), width=9, bg="sky blue")\
                        .grid(row=0, column=j)

                    machine_no = int(machine_efficiency_value)
                    med_filter = stock[
                        (main_csv['Center Name'] == centres[i]) & (main_csv['Service Sub Group'] == 'Dialysis') & (
                                main_csv['Bill Open Date'] == str(date))]

                    if j == 0:
                        lbl_centre = tk.Label(frame_buttons_2, text=centres[i], width=35, font=("Arial", 8))
                        lbl_centre.grid(row=i + 1, column=j)
                        tk.Label(frame_buttons_2, text='Centre', font=("Arial", 8), width=9, bg="sky blue")\
                            .grid(row=0, column=0)
                    elif j == j:
                        lbl_value_id = tk.Label(frame_buttons_2, text=round((len(med_filter['Bill Open Date']) /
                                                                             (machine_no * 3)) * 100, 2), width=10)
                        lbl_value_id.grid(row=i + 1, column=j)

                    grand_total = grand_total + (len(med_filter['Bill Open Date']) / (machine_no * 3)) * 100
            except:
                pass
    frame_buttons_2.update_idletasks()

    # Set the canvas scrolling region
    canvas_2.config(scrollregion=canvas_2.bbox("all"))

    # ######################### Daily Business Report CSV Validation ###########################################

    try:
        if main_csv == 0 or main_csv.empty:
            root_1.destroy()
            messagebox.showerror("Error ", "Please import daily Business report csv  file ")

            return False
    except:
        pass

    #############################################################################################

    # ################################### Machine_Validation #####################################################
    try:
        if machine_efficiency == 0 or machine_efficiency.empty:
            messagebox.showerror("Error", "Please import machine_efficiency file ")
            return False
    except:
        pass
    ##########################################################################################

    # ############################################### Selection Validation ########################
    if len(selected_value) == 0:
        root_1.destroy()
        messagebox.showerror('Error', 'Please select the centers')

        return False
    else:
        return True
    # Launch the GUI

    root_1.mainloop()


tk.Button(root, text="Click to Generate MIS", highlightcolor='red', width=70, height=2, bg="white", fg="blue",
          command=show_output).place(x=150, y=630)
root.mainloop()
```
